```python
from transformers import GPT2Config 
from transformers import GPT2LMHeadModel 
model_name_or_path = "/home/pkunwar/characterize_ttlora/gpt2small/checkpoints"
cache_dir = "/home/pkunwar/.cache/huggingface/transformers"
```

```python
config = GPT2Config.from_pretrained(model_name_or_path, cache_dir=cache_dir)
gpt2 = GPT2LMHeadModel.from_pretrained(
        model_name_or_path,
        config=config,
        cache_dir=cache_dir,
    )
print(f'base gpt2 model: {model_name_or_path}')
print(gpt2)
```
